```
import numpy as np
import matplotlib.pyplot as plt

# --- 1. DATOS DE LOS EJERCICIOS ---
sumas = [
    (np.array([5, 8]), np.array([1, 2])),
    (np.array([-1, 0]), np.array([2, 3])),
    (np.array([0.5, 1]), np.array([1.5, 2])),
    (np.array([10, 20]), np.array([5, 5])),
    (np.array([1, 2]), np.array([0, 0]))
]

restas = [
    (np.array([20, 10]), np.array([10, 5])),
    (np.array([1, 1]), np.array([0.5, 0.5])),
    (np.array([-5, 2]), np.array([2, -1])),
    (np.array([100, 50]), np.array([80, 40])),
    (np.array([0, 0]), np.array([1, 2]))
]

escalares = [
    (3, np.array([2, 4])),
    (0.1, np.array([100, 50])),
    (-2, np.array([1, -2])),
    (0.5, np.array([4, 8])),
    (10, np.array([1.2, 0.5]))
]

# --- 2. EJECUCIÓN Y CÁLCULOS ---
print("=== VALIDACIÓN DE OPERACIONES VECTORIALES ===\n")

print("--- SECCIÓN: SUMAS ---")
for i, (u, v) in enumerate(sumas, 1):
    print(f"{i}. {u} + {v} = {u + v}")

print("\n--- SECCIÓN: RESTAS ---")
for i, (u, v) in enumerate(restas, 1):
    print(f"{i}. {u} - {v} = {u - v}")

print("\n--- SECCIÓN: ESCALARES ---")
for i, (k, u) in enumerate(escalares, 1):
    print(f"{i}. {k} * {u} = {k * u}")

# --- 3. GRAFICACIÓN DEL CASO REAL (CINEMÁTICA) ---
v_avion = np.array([0, 500])
v_viento = np.array([100, -50])
v_resultante = v_avion + v_viento

plt.figure(figsize=(7, 7))
plt.quiver(0, 0, v_avion[0], v_avion[1], color='b', scale=1, scale_units='xy', angles='xy', label='Avión')
plt.quiver(v_avion[0], v_avion[1], v_viento[0], v_viento[1], color='r', scale=1, scale_units='xy', angles='xy', label='Viento')
plt.quiver(0, 0, v_resultante[0], v_resultante[1], color='g', scale=1, scale_units='xy', angles='xy', label='Resultante')

plt.xlim(-50, 200)
plt.ylim(-100, 600)
plt.grid(True)
plt.legend()
plt.title("Validación Cinemática: Navegación Aérea")
plt.savefig('evidencia_vectores.png')
print("\n[INFO] Imagen 'evidencia_vectores.png' generada con éxito.")
plt.show()
```
